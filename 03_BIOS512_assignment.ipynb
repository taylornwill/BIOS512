Make a histogram or KDE plot that shows the distribution of body mass in each year. *Use a different theme than the default.*

```r
library('scales')
```

```r
options(repr.plot.width = 6, repr.plot.height = 6)
p = ggplot(penguins, aes(x = body_mass_g)) + geom_histogram(fill = 'blue', color = 'white',binwidth = 100) +
    facet_wrap(~year, ncol = 1) + 
    labs(title = 'Histograms of Penguin body mass for measurements taken from 2007-2009', 
        x = 'Body mass in grams', y = 'Number of Penguins') +
    theme_minimal() + 
    theme(strip.text = element_text(hjust = 0.9), plot.title = element_text(size = 11)) + 
    scale_x_continuous(label = comma)  

p
```

Is there a difference between the body mass of male and female penguins? Use a scatter plot with jitter to compare the masses of male and female penguins. **Facet by `species` to compare within each species-group.**

```r
options(repr.plot.height = 7, repr.plot.width = 5)
p = ggplot(penguins, aes(x = body_mass_g, y = sex)) + 
    geom_jitter(alpha = 0.5, width = 0.4, height = 0.2) + 
    facet_wrap(~species, ncol = 1) +
    labs(x = 'Body Mass in grams',y = NULL) + 
    theme_minimal() + theme(strip.text = element_text(hjust = 0)) +
    scale_x_continuous(labels = comma)
p
```
